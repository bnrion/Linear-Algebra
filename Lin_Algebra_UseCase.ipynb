### Representasi Data

#### Teks
Dokumen atau kalimat dapat direpresentasikan sebagai matriks yang mengandung frekuensi kemunculan kata-kata.

Dalam pemrosesan bahasa alami, representasi dokumen atau kalimat dapat direpresentasikan sebagai matriks yang mengandung frekuensi kemunculan kata-kata. Representasi ini dikenal dengan istilah bag-of-words.

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = ["Ini adalah dokumen pertama.", "Ini adalah dokumen kedua.", "Ini adalah kalimat pertama."]

vectorizer = CountVectorizer()

matrix = vectorizer.fit_transform(documents)

matrix
```

### Linear Regresi

Regresi linier: Regresi linier merupakan model machine learning sederhana yang dapat digunakan untuk memprediksi nilai numerik. Regresi linier memanfaatkan operasi matriks, khususnya perkalian matriks, untuk menghitung prediksi berdasarkan bobot dan fitur dari data.

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

X = 2.5 * np.random.rand(100) + 1.5 #luas bangunan # variabel independen
```

```python
X
```

```python
res = 0.2 * np.random.randn(100)  # random error -> nilai bias
res
```

```python
y = 0.3 * X + res #harga rumah # variabel dependen

#rumus linear regresi -> simple linear regresi
# y = a * X + error

# mengubah variabel independen menjadi matriks
X_matrix = np.column_stack((X, np.ones(len(X))))
X_matrix
```

```python
# menghitung koefisien regresi
beta = np.linalg.inv(X_matrix.T.dot(X_matrix)).dot(X_matrix.T).dot(y)
```

```python
# menampilkan hasil regresi
print("Koefisien regresi: ", beta)
```

```python
# plot data dan hasil regresi
plt.scatter(X, y)
plt.plot(X, X_matrix.dot(beta), color='red')
plt.show()
```
